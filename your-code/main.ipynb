####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
emsc = requests.get(url)
soup_emsc = BeautifulSoup(emsc.content, 'lxml')
````

````python
import pandas as pd

emsc_quakes = soup_emsc.select('tbody#tbody>tr:nth-of-type(-n+20)')

emsc_quakes_date_and_time = [i.text for x in emsc_quakes for i in x.select('b>a')]
emsc_quakes_date_and_time = [i.split('\xa0\xa0\xa0') for i in emsc_quakes_date_and_time]

emsc_quakes_latn_and_lonn = [re.sub('\xa0', '',i.text) for x in emsc_quakes for i in x.select('td.tabev1')]
emsc_quakes_latn = emsc_quakes_latn_and_lonn[0::2]
emsc_quakes_lonn = emsc_quakes_latn_and_lonn[1::2]

emsc_quakes_latc_and_lonc = [re.sub('\xa0', '',i.text) for x in emsc_quakes for i in x.select('td.tabev2')]
emsc_quakes_latc = emsc_quakes_latc_and_lonc[0::3]
emsc_quakes_lonc = emsc_quakes_latc_and_lonc[1::3]

emsc_quakes_region = [re.sub('\xa0', '',i.text) for x in emsc_quakes for i in x.select('td.tb_region')]

df_quakes = pd.DataFrame(zip(list(zip(*emsc_quakes_date_and_time))[0],list(zip(*emsc_quakes_date_and_time))[1], emsc_quakes_latn, emsc_quakes_latc, emsc_quakes_lonn, emsc_quakes_lonc, emsc_quakes_region),
columns = ['Date','Time','Lat','LatC','Lon','LonC','Region'])
df_quakes
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
wiki = requests.get(url)
soup_wiki = BeautifulSoup(wiki.content, 'lxml')

languages = soup_wiki.select('div.central-featured>div>a>strong')
languages_list = [i.text for i in languages]

articles = soup_wiki.select('div.central-featured>div>a>small>bdi')
articles_list = [re.sub('\xa0', '',i.text) for i in articles]

df_wiki = pd.DataFrame(zip(languages_list, articles_list), columns=['Language','Articles'])
df_wiki
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
native = requests.get(url)
soup_native = BeautifulSoup(native.content, 'lxml')
lan_nat = soup_native.select('div.mw-parser-output>table:nth-of-type(2)>caption~tbody>tr>td:nth-child(-n+3)')

lan_nat_data = [re.sub('\[\d+\]','',re.sub('\n','',i.text)) for i in lan_nat]
rank = lan_nat_data[0::3]
lan = lan_nat_data[1::3]
speak = lan_nat_data[2::3]
df_lan_nat = pd.DataFrame(zip(rank, lan, speak), columns=['Ranking','Language','Speakers (M)'])
df_lan_nat
````
